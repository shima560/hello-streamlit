## if文
　「if」は記述した条件に対してTrueかFalseかを計算し, Trueの場合のみ実行する構文.  複数条件を順番に確認するときは「elif」(else if), どの条件にも当てはまらなかったときは「else」を使い, 下記の様にその下の部分が実行される.  
>```
if 条件1:  
    条件1がTrueのときに実行する処理  
elif 条件2: 
    条件1がFalseで条件2がTrueのときに実行する処理  
elif 条件3:  
    条件1, 2がFalseで条件3がTrueのときに実行する処理  
else:  
    全ての条件がFalseのときに実行する処理  

※if, elif, elseの下の実行する処理部分は**左側に適切な長さのインデント(空白)が必要です**→ 半角スペース4つ分か, tabキー1つ分

````python
a = 10
````

````python
if a > 0:
    # str関数でaを文字型に変換
    print(str(a) + "は0より大きい")
````

````python
a = 3
````

````python
if a > 0:
    print(str(a) + "は0より大きい")
elif a == 0:
    print(str(a) + "は0")
else:
    print(str(a) + "は0未満")
````

## for文
　for文とは繰り返し文のことで, 順番に値を変数に格納しながら繰り返し実行する構文です. 複数の要素からなるオブジェクトの要素を順番に指定した変数に格納して, 実行します.
>```
for i in リスト:
    繰り返し実行する処理

　iに, 順番にリストの中身が格納されて「繰り返し実行する処理」が実行されます. リストの最後の要素がiに格納されて「繰り返し実行する処理」が実行されたら繰り返しは終了です. また, range関数というものがあり, 0から連続する整数を簡単に生成してくれる関数です.  
  
※参考：range関数は引数を用いるとstart, stop, stepを制御できます

````python
for i in ["Sano", "Ryotaro", "です"]:
    print(i)
````

````python
for i in range(10):
    if i % 3 == 0:
        print(str(i) + "は3の倍数です")
    else:
        print(str(i) + "は3の倍数ではないです")
````

### オブジェクトに紐づく関数
　<span style="color: red; ">ここでは簡易的な説明だけをします</span>. オブジェクトに紐づく関数をメソッドと呼びます. 通常は$f(x)=2x$と関数$f$を定義すると$x$には**数値**しか入れることができません, 例えば$x$に文字列が代入されていると$f(x)$が計算ができないからです.  
　つまり, 関数とはオブジェクト(変数)に紐づく訳です.  
　→　オブジェクトが決まると, 使用できる関数も決まってくる  
　→　数値には数値用の関数, 文字列には文字列用の関数  

　例えば, dataframe型のオブジェクトにはデータの先頭数行だけを抽出する関数「head」が紐づいています. このとき, dataframe型のオブジェクトをdatとすると, **head関数の使い方はhead(dat)でなくdat.head()とオブジェクトに紐づいて書きます**.  

※head関数はパラメータを1つもっており, 先頭から何行抽出するかも指定できます.

````python
dat.head()
````

````python
dat.head(2)
````

````python
# この書き方ではエラーが出ます
pd.head(dat)
````

　**ここでのポイント**はpd.read_csv()のようにオブジェクトに紐づかない関数と, head()のようにオブジェクトに紐づく関数があり, 書き方に大きく2通りあるということを覚えてください.
